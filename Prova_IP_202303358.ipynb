No trabalho em grupo que estamos desenvolvendos, o principal framework que estamos utilizando é o [PyCaret](https://pycaret.org/), esse framework é um conglomerado de outros framworks, Ray, Prophet, CatBoost, LightGBM e etc. Ele é um framework open-source e low-code voltado para a area de machine learning, com o intuito de tornar essa area mais acessivel para quem esta começando a aprender sobre o assunto(Oque eles chamam de Democratização do Aprendizado de Máquina), por isso ele é low-code de fácil uso.


O PyCaret possui a vantagem de realizar coisas complexas em poucas linhas de código pois ele é muito automatizado, especificamente, ele é bem eficiente para [algoritmos](https://pycaret.gitbook.io/docs/get-started/modules) de Classificação, Regressão, Clusterização, Detecção de Anomalias e Serie Temporal. Além disso, dependendo do problema a ser resolvido, ele não será tão eficiente, pois ele não é tão especifico e otimizado para uma tarefa em si.

Nesse sentido, podemos concluir que o PyCaret possui muitas aplicações em diversas areas, como detecção de fraudes em sistemas bancarios, detecção de problemas em exames medicos, previsões de evasão de clientes, predição de informaçãoes como vendas, temperatura , realização de forecasting por serie temporal e etc.

surgiu, você explicou o Numpy, durante uma partida de jogo da velha. Ele fez uma

você aceitou o desafio e apresentou este código, explicando.

```python
def verify_game(velha): # Essa função verifica se houve ganhador
  #Caso ocorra uma vitoria, a função return um True para que o codigo entenda que deva parar
  for linha in velha: #Verifica se houve vitoria nas linhas
    if linha[0] == linha[1] == linha[2] != 0: 
      print(f'O jogador {linha[0]:.0f} venceu! :)')
      return True

  for coluna in range(3): #Verifica se houve vitoria nas colunas
    if velha[0][coluna] == velha[1][coluna] == velha[2][coluna] != 0: 
      print(f'O jogador {velha[0][coluna]:.0f} venceu! :)')
      return True

  if velha[0][0] == velha[1][1] == velha[2][2] != 0: #Verifica se houve vitoria em uma diagonal
    print(f'O jogador {velha[0][0]:.0f} venceu! :)')
    return True

  if velha[0][2] == velha[1][1] == velha[2][0] != 0: #Verifica se houve vitoria na outra diagonal
    print( f'O jogador {velha[0][2]:.0f} venceu! :)')
    return True

def letrinha(velha,i,j):
  if velha[i][j] == 1: #Se o valor corresponder a 1, ele susbtitui por X no tabuleiro
    return 'X'
  elif velha[i][j] == 2: #Se o valor corresponder a 2, ele susbtitui por O no tabuleiro
    return '◯'
  else: #Caso nenhuma coisa ocorra, ele apenas deixa um espaço em branco
    return ' '


def tabuleiro(velha): #Mostra no output, o tabuleiro de uma forma bonitinha
  print(f' {letrinha(velha,0 ,0)} │ {letrinha(velha,0,1)} │ {letrinha(velha,0,2)}', flush=True)
  print('───┼───┼───', flush=True)
  print(f' {letrinha(velha,1,0)} │ {letrinha(velha,1,1)} │ {letrinha(velha,1,2)}', flush=True)
  print('───┼───┼───', flush=True)
  print(f' {letrinha(velha,2,0)} │ {letrinha(velha,2,1)} │ {letrinha(velha,2,2)}', flush=True)

def jogada(numero, letra, velha): #Função que faz os principais movimentos das jogadas
  while True:
    escolha = input(f'Jogador {numero}, escolha uma posição para colocar o {letra}: ')
    escolha = list(map(int, escolha.replace(' ', '').replace(',','')))
    #Recebe os valores inputados pelo player

    if (escolha[0] >= 4) or (escolha[1] >= 4): #Verifica o numero possui uma posição no tabuleiro
      print('O numero digitado não possui uma posição no tabuleiro!', flush=True)
    else:
      if velha[escolha[0]-1][escolha[1]-1] == 0:
        velha[escolha[0]-1][escolha[1]-1] = numero
        break #verifica se o player pode ocupar a posição desejada
      else:
        print(f'Posição ja ocupada!\nDigite outra posição para colcoar o {letra}', flush=True)
  tabuleiro(velha)



vencedor = 0
#Variavel que verifica se houve vencedor, parece ver se ha ou não empate

velha = np.zeros((3,3))
#Cria a matriz que serve de base para o jogo


print('-'*5+' Bem vindo ao Jogo da Velha do BIA '+'-'*5+'\n')
print('Este jogo é em dupla, então chame o seu amigo! :)')
print('Para colocar a sua posicão desejada, digite o numero')
print('da coluna, depois o numero da linha, da seguinte maneira')
print('reposta: <numero da linha> <numero da coluna>\n')
print('         Col1:  Col2:  Col3:')
print('Linha 1:      │      │     ')
print('         ─────┼──────┼─────')
print('Linha 2:      │      │     ')
print('         ─────┼──────┼─────')
print('Linha 3:      │      │     \n')


#Codigo que roda o jogo em si
try:
  for i in range(9): #São nove vezes que são jogados ao maximo
    if (i%2) == 0:
      jogada(1, 'X', velha)
      if verify_game(velha):
        vencedor = 1
        break #Fecha o game caso haja um vencedor
    else:
      jogada(2, 'O', velha)
      if verify_game(velha):
        vencedor = 1
        break #Fecha o game caso haja um vencedor
    clear_output(wait=True)
  if vencedor == 0: #Verifica se deu empate
    print("Deu empate!")

except Exception as error: #tratamento de erros
  print(f'Ocorreu um erro! :(\n{error}')
```

3. (valor 2,0) Você está participando de um projeto de pesquisa do CEIA, em parceria
com um fabricante de sensores do tipo sonar. Sua tarefa é desenvolver um
aplicativo em Python que calcula a distância até um objeto detectado pelo sonar. O
sensor sonar funciona emitindo um sinal e medindo o tempo que leva para o sinal
retornar, após ser refletido pelo objeto. Este tempo é então enviado para o seu
aplicativo. A velocidade de propagação do som na água varia entre 1.450 e 1.570 metros por
segundo. Com base nisso, escreva um programa em Python que receba o tempo de retorno do sinal
e calcule a distância média até o objeto.

```python
def calculo_distancia(tempo): #Função que faz os calculos de distancia e mostra ao usuario
    distancia_min = (tempo/2)*1450
    distancia_max = (tempo/2)*1570
    media_distancia = (distancia_max+distancia_min)/2
    #A velocidade de propagação da agua varia entre 1450m/s e 1570m/s
    #Por isso trabalharemos com duas contas para termos um range entre as possiveis distancias
    #O sinal vai é volta, por divimos o sinal por 2 para desconsiderarmos isso
    #Apos isso, e so realizar a conta de cinemática básica
    
    print(f'A distancia entre o sonar e o objeto é de aproxidamente {media_distancia}m')
    print(f'Entretanto, por variações normais da agua, o objeto pode estar entre {distancia_min}m e {distancia_max}m' )

while True:
    try:
        print('Digite 1 caso você queria colocar o tempo de retorno do sinal manualmente, ou')
        print('Digite 2 caso você queria que o tempo fornecido seja aleatorio')
        sinal = int(input('>> '))
        clear_output(wait=True)
        if sinal == 1:
            tempo = float(input('Tempo de retorno do sinal(s): '))
            #A variavel tempo recebe o tempo entre a emissão e o retorno do sinal do sonar.

            print(f'O tempo fornecido é de {tempo}s')
            calculo_distancia(tempo)
            break
        elif sinal == 2:
            tempo =  random.randrange(1,100)
            #Tempo fornecido é aleatorio, em uma situação, isso é algo que não se queira
            #Mas para efeito didaticos, é eficiente
            
            print(f'O tempo gerado aleatoriamente é {tempo}s')
            calculo_distancia(tempo)
            break
        else:
            print('Digite uma opção viavel')
    except:
        pass
```
